```python
#import pandas, requests, json

import pandas as pd
import requests
import json
import numpy as np
import csv
import re
```

```python
#telling system where api key is located
with open('APIkeys.json') as f:
    keys = json.load(f)
    secretkey = keys['secretapikey']
```

```python
#running the code using 3 different lat/lon for project
url = "https://trailapi-trailapi.p.rapidapi.com/trails/explore/"
querystring1 = {"lat":"47","lon":"-122"}
querystring2 = {"lat":"28","lon":"-81"}
querystring3 = {"lat":"44","lon":"-72"}
headers = {
    "X-RapidAPI-Key": secretkey,
    "X-RapidAPI-Host": "trailapi-trailapi.p.rapidapi.com"
}
response1 = requests.get(url, headers=headers, params=querystring1)
print(response1.json())
```

```python
#making sure to print all results to verify data
response2 = requests.get(url, headers=headers, params=querystring2)
response3 = requests.get(url, headers=headers, params=querystring3)
print(response2.json())
```

```python
#Step 3
#converting dataframe to csv
with open('api4.csv','a') as f:
    for df in df4:
        df.to_csv(f)
        f.write("\n")
```

```python
#reading csv file as a df
df=pd.read_csv("api4.csv")
df.head(10)
```

```python
#verifying that all files were combined by looking at the different lat/lon
#Step 4
df.tail()
```

```python
#Step 5
#finding out the data types
df.dtypes
```

```python
#finding NaN values
for c in df.columns:
    nan = df[c].isnull().values.any()
    #printing out 
    if nan == True:
        print(" {} has Nan value(s)".format(c)) 
    else: 
        print("{} has NO Nan values!".format(c))
```

```python
#removing all nan values from lat/lon
#step 6
df = df.dropna(subset = ['lat'])
df = df.dropna(subset = ['lon'])
```

```python
#changing type of lat/lon to numeric, ignoring any errors due to the negative values
df['lat'] = pd.to_numeric(df['lat'], errors='coerce')
df['lon'] = pd.to_numeric(df['lon'], errors='coerce')
#verified change
print(df.dtypes)
```

```python
#rounding values to make them easier to connect on the other tables
#Step 7
df['lat'].round()
```

```python
#rounding values to make them easier to connect on the other tables
#step 7
df['lon'].round()
```

This data was pulled via an API that finds trails from an entered latitude and longitude. This data was transformed from json to a dataframe. It's interesting because, even though a lat and lon had to be entered, there were still nan values in these fields on occasion, which had to be removed. ALso, I rounded all of the lat/lon values to whole numebers. This is because it will be very difficult to match up lat/lon depending on where the different sources used the data to get those points from, especially if the data goes to the 3rd or 4th decimal. This could cause issues in just how close a specific trail might be to someone. It might show someone a trail that is much farther than they were hoping for. However, since there are so many options to choose from, hopefully it'll be fairly easy to simply find a new trail. There is definitely room for error in such a simplified version of this project. But, this data could be made more accurate, if the the code was a bit more complicated.
