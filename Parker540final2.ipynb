```python
#read the csv file from project in a dataframe
df=pd.read_csv("countries.csv")
```

```python
#Step 1
#finding duplicate values
for c in df.columns:
    
    dup = any(df[c].duplicated())
    
    if dup == True:
        print(" {} has duplicate value(s)".format(c)) 
    else: 
        print("{} has NO  duplicate values!".format(c))
```

```python
#Step 2
#finding NaN values
for c in df.columns:
    nan = df[c].isnull().values.any()
    #printing out 
    if nan == True:
        print(" {} has Nan value(s)".format(c)) 
    else: 
        print("{} has NO Nan values!".format(c))
```

```python
#removed capitals with nan values
df1 = df.dropna(subset = ['capital'])
```

```python
#checked that removal was succesful! 
print("The column Capital contains NaN - %s " % df1.capital.isnull().values.any())
```

```python
#Step 3
#rename headers
df1.rename(columns={"iso3": "iso_3"}, inplace=True)
df1.rename(columns={"iso2": "iso_2"}, inplace=True)
df1.rename(columns={"emojiU": "emoji_u"}, inplace=True)
df1.rename(columns={"timezones": "time_zones"}, inplace=True)
```

```python
#Step 4
#removed some columns that aren't needed
df2 = df1[['id', 'name', 'iso_3', 'iso_2', 'phone_code', 'capital', 'currency', 'currency_name', 'currency_symbol', 'native', 'region', 'subregion', 'latitude', 'longitude']].copy()
```
